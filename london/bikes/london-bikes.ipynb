```python
%matplotlib inline

import colorsys
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
from bokeh.plotting import figure, show
from bokeh.resources import CDN
from bokeh.io import output_notebook
output_notebook( resources=CDN )

pd.set_option('max_colwidth', 200)
```

```python
rawData       = pd.read_csv('london-bikes-coords.csv', index_col=0)
```

```python
rawData
```

```python
graph = nx.Graph()
```

```python
uniqueLocations = {}
uniqueCount = 0

for id, info in rawData.iterrows():
    if info["StartStation Name"] not in uniqueLocations:
        name = info["StartStation Name"]
        lat = info["StartLatitude"]
        long = info["StartLongitude"]
        uniqueLocations[name] = (lat, long)
        uniqueCount += 1
        
    if info["EndStation Name"] not in uniqueLocations:
        name = info["EndStation Name"]
        lat = info["EndLatitude"]
        long = info["EndLongitude"]
        uniqueLocations[name] = (lat, long)
        uniqueCount += 1
    
```

```python
print(uniqueCount)
```

```python
# list of unique locations and their coordinates
pd.DataFrame.from_dict(uniqueLocations, orient="index", columns=['latitude', 'longitude'])
```
